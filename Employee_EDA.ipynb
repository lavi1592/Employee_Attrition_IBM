# Level-1 Analysis

```python
def separate_data_types(sf):
    categorical=[]
    numerical=[]
    for i in sf.columns:
        if sf[i].nunique()<40:
            categorical.append(i)
        else:
            numerical.append(i)
    return categorical, numerical

categorical, numerical = separate_data_types(df)            
```

```python
from tabulate import tabulate 
```

```python
table=[categorical, numerical]
print(tabulate({'categorical':categorical, 'numerical': numerical}, headers=['Categorical','Numerical']))
```

## Univariate Analysis of Categorical Variables

```python
def info_of_cat(col): 
    print(f"Unique values in {col} are: {df[col].unique()}") # unique: returns the unique values in the col
    print(f"Mode of {col} is {df[col].mode()[0]}")  # mode: returns the mode of the column
    print(f"Number of missing values in {col} is {df[col].isnull().sum()}") # isnull().sum() counts the number of null in dataframe
```

### 1. Attrition

```python
info_of_cat("Attrition")
```

```python
fig, ax=plt.subplots(1,2, figsize=(15,7))
ax[0].set_title("Count of Attrition")
ax[1].set_title("Percentage share of Attrition")
data= df['Attrition'].value_counts()
labels= data.keys()

sns.countplot(x= df['Attrition'], ax=ax[0])
plt.pie(x=data, autopct="%.1f%%", explode=[0.02,0.1], labels=labels, pctdistance=0.6)
plt.show()
```

## Interpretation: Count of employees who are still working in the company are more compared to the employees who left the company
